# Wall

```python
image_rgb = cv2.imread('wall1.jpg')
image_rgb = cv2.cvtColor(image_rgb,cv2.COLOR_BGR2RGB)
#plt.imshow(image_rgb)

image_gray = cv2.imread('wall1.jpg',0)
#plt.imshow(image_gray,cmap='gray')

edges = cv2.Canny(image_gray,threshold1=0,threshold2=100)

blurred_image = cv2.blur(image_gray,ksize=(15,15))
edges2 = cv2.Canny(blurred_image,threshold1=0,threshold2=100)

fig, axs = plt.subplots(2,2,figsize=(10,10))
axs[0,0].imshow(image_gray,cmap='gray')
axs[0,1].imshow(edges,cmap='gray')
axs[1,0].imshow(blurred_image,cmap='gray')
axs[1,1].imshow(edges2,cmap='gray')
```

# Road

```python
image_rgb = cv2.imread('road1.jpg')
image_rgb = cv2.cvtColor(image_rgb,cv2.COLOR_BGR2RGB)
#plt.imshow(image_rgb)

image_gray = cv2.imread('road1.jpg',0)
#plt.imshow(image_gray,cmap='gray')

median_value = np.median(image_gray)
thresh1 = int(max(0,0.7*median_value))
thresh2 = int(min(255,1.3*median_value))

edges = cv2.Canny(image_gray,threshold1=thresh1,threshold2=thresh2)

blurred_image = cv2.blur(image_gray,ksize=(5,5))
edges2 = cv2.Canny(blurred_image,threshold1=thresh1,threshold2=thresh2)

fig, axs = plt.subplots(2,2,figsize=(20,10))
axs[0,0].imshow(image_gray,cmap='gray')
axs[0,1].imshow(edges,cmap='gray')
axs[1,0].imshow(blurred_image,cmap='gray')
axs[1,1].imshow(edges2,cmap='gray')
```
