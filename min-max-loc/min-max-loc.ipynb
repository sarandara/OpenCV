# Min-Max-Loc

```python
import cv2
import matplotlib.pyplot as  plt
%matplotlib inline
import numpy as np
```

Image Source: https://unsplash.com/photos/9obtGmpfglg

```python
comet = cv2.imread('comet.jpg')
comet = cv2.cvtColor(comet,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(comet)
```

```python
comet_gray = cv2.cvtColor(comet,cv2.COLOR_BGR2GRAY)
```

```python
comet_gray.shape
```

```python
plt.imshow(comet_gray,cmap='gray')
```

```python
cv2.minMaxLoc(comet_gray)
```

```python
comet_gray2 = np.copy(comet_gray)
```

```python
max_location = cv2.minMaxLoc(comet_gray)[3]
```

```python
max_location
```

```python
max_location_x = max_location[0]
max_location_y = max_location[1]
```

```python
color = (255,0,0)
t = 2
```

```python
vline1 = (max_location_x,max_location_y-100)
vline2 = (max_location_x,max_location_y+100)
hline1 = (max_location_x-100,max_location_y)
hline2 = (max_location_x+100,max_location_y)
```

```python
comet_2 = np.copy(comet)
comet_loc = cv2.line(comet_2,vline1,vline2,color,t)
comet_loc = cv2.line(comet_2,hline1,hline2,color,t)
```
