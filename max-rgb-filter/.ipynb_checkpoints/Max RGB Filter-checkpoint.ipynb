# Max RGB Filter

# Import

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

SOURCE: https://unsplash.com/photos/Hj9M7A4FVTc

```python
filename = 'crater.jpg'
```

```python
img = cv2.imread(filename)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_new = img.copy()
```

```python
r = img_new[:,:,0]
g = img_new[:,:,1]
b = img_new[:,:,2]
```

```python
max_rgb = cv2.max(cv2.max(r,g),b)
```

```python
r[r<max_rgb]=0
g[g<max_rgb]=0
b[b<max_rgb]=0
```

```python
plt.imshow(img_new)
```

```python
fig, axs = plt.subplots(1,2,figsize=(20,10))
axs[0].imshow(img)
axs[0].set_title('pre');
axs[1].imshow(img_new)
axs[1].set_title('post');
```

```python
fig.savefig('crater-rgb-max.jpg')
```
