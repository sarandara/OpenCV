# Saving Images

## Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

## Load

IMAGE SOURCE: https://unsplash.com/photos/cdjsw6sjwLY

```python
img = cv2.imread('comet-original.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img,cmap='gray')
```

## Save Method - Matplotlib

```python
plt.imsave('comet-gray-1.jpg',img, cmap='gray')
```

## Save Method - Matplotlib, with axes

```python
img_and_axes = plt.imshow(img,cmap='gray')
```

```python
img_and_axes = img_and_axes.get_figure()
```

```python
img_and_axes.savefig('comet-gray-1-with-axes.jpg')
```

## Save Method - PIL

```python
img_pil = Image.fromarray(img)
```

```python
img_pil.save('comet-gray-2.jpg')
```

## Save Method - OpenCV

```python
cv2.imwrite('comet-gray-3.jpg',img)
```
